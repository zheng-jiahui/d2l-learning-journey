```python
import torch
from torch import nn
from d2l import torch as d2l
```

```python
def corr2d(X, K):
    """计算二维互相关运算"""
    h, w = K.shape
    Y = torch.zeros((X.shape[0]-h+1, X.shape[1]-w+1))
    for i in range(Y.shape[0]):
        for j in range(Y.shape[1]):
            Y[i, j] = (X[i:i+h, j:j+w] * K).sum()
    return Y
```

```python
K = torch.tensor([[2.0, -2.0]])
X = torch.ones((6, 8))
X[:, 2:6] = 0
Y = corr2d(X, K)
print(Y)
```

```python
conv2d = nn.Conv2d(1, 1, kernel_size=(1,2), bias=False)

X = X.reshape((1, 1, 6, 8))
Y = Y.reshape((1, 1, 6, 7))
lr = 3e-2  # 学习率

for i in range(10):
    y_hat = conv2d(X)
    l = (y_hat - Y) ** 2
    conv2d.zero_grad()
    l.sum().backward()
    # 手动更新参数
    conv2d.weight.data -= lr * conv2d.weight.grad
    print(f'epoch {i + 1}, loss {l.sum():.3f}, weight {conv2d.weight.data.numpy()}')
```
